```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_path = '/content/drive/MyDrive/Colab Notebooks/이어드림스쿨/기업강의(fastcampus)/인공지능 머신러닝(0530)/실습예제/data/'

train = pd.read_csv(base_path + 'train.csv')
test = pd.read_csv(base_path + 'test.csv')
```

```python
train.head()
```

```python
## 아웃라이어 제거

outliers = train[train['y'] > 135]
train = train.drop(index=outliers.index)
train
```

```python
## categorycal variables --> One-Hot encoding

cat_vars = train.columns[2:10]
binary_vars = train.columns[10:]

display(cat_vars, binary_vars)
```

```python
## Encoding Categorical features

## 1. Ordinal Encoding
for var in cat_vars:
    train[var] = pd.factorize(train[var])[0]
```

```python
train.head()
```

```python
## binary col중 모모든 데이터가 0인 경우 제거해준다.

all_0_cols = []

for col in binary_vars:
    one_percent = train[col].mean() * 100
    if one_percent == 0.0:
        all_0_cols.append(col)

all_0_cols
```

```python
train = train.drop(columns=all_0_cols)
train
```

---

## feature engineering
    1. Correlation
    2. VIF
    3. PCA

```python
## correlation이 높은 변수 찾기

temp_vif = train.drop(columns=['ID', 'y'])
temp_pca = train.drop(columns=['ID', 'y'])
temp_vif.shape, temp_pca.shape
```

```python
## VIF 가 10이 넘는 변수 찾기
## VIF_i = 1 / 1 - R2_i

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_list = [variance_inflation_factor(temp_vif.values, idx) for idx in range(len(temp_vif.columns))]
vif_df = pd.DataFrame(index=temp_vif.columns, data=vif_list, columns=['VIF'])
vif_df.sort_values(by='VIF', ascending=False)
```

```python
vif_cols = vif_df[(vif_df['VIF'] > 10)].index
```

```python
temp_vif = temp_vif.drop(columns=vif_cols)
```

```python
vif_list = [variance_inflation_factor(temp_vif.values, idx) for idx in range(len(temp_vif.columns))]
vif_df = pd.DataFrame(index=temp_vif.columns, data=vif_list, columns=['VIF'])
vif_df.sort_values(by='VIF', ascending=False)
```

```python
vif_cols = vif_df[(vif_df['VIF'] > 10)].index
temp_vif = temp_vif.drop(columns=vif_cols)
```

```python
vif_list = [variance_inflation_factor(temp_vif.values, idx) for idx in range(len(temp_vif.columns))]
vif_df = pd.DataFrame(index=temp_vif.columns, data=vif_list, columns=['VIF'])
vif_df.sort_values(by='VIF', ascending=False)
```

```python
## 이제 VIF의 값이 10 이상이 column이 없다.

vif_df.sort_values(by='VIF', ascending=False)
```

## PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X = pca.fit_transform(temp_pca)

temp_pca.shape, X.shape
```

```python
pca_columns = [f"PC{i}" for i in range(X.shape[1])]
pca_df = pd.DataFrame(columns=pca_columns, data=X)
pca_df
```

```python
## 4. 학습데이터 분할 (vif)

from sklearn.model_selection import train_test_split

X = temp_vif
y = train['y']
X_2 = pca_df
y_2 = train['y']

X.shape, y.shape, X_2.shape, y_2.shape
```

```python
X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.2, random_state=100)

X_train_2, X_test_2, y_train_2, y_test_2 = \
train_test_split(X_2, y_2, test_size=0.2, random_state=100)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

from lightgbm.sklearn import LGBMRegressor
```

```python
param_grid = {
    "max_bin" : 20,
    "learning_rate" : 0.0025,
    "objective" : "regression",
    "boosting_type" : "gbdt",
    "metric" : "mae",
    "sub_feature" : 0.345,
    "bagging_fraction" : 0.85,
    "bagging_freq" : 40,
    "num_leaves" : 512,
    "min_data" : 500,
    "min_hessian" : 0.05,
    "verbose" : 2,
    "feature_fraction_seed" : 2,
    "bagging_seed" : 3
}

model = LGBMRegressor(**param_grid)
model_2 = LGBMRegressor(**param_grid)
```

```python
print("\nFitting LightGBM...")
model.fit(X_train, y_train)
model_2.fit(X_train_2, y_train_2)
```

```python
from sklearn.metrics import r2_score

```

```python
print("Prediction")
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

pred_train_2 = model_2.predict(X_train_2)
pred_test_2 = model_2.predict(X_test_2)


train_score = r2_score(y_train, pred_train)
test_score = r2_score(y_test, pred_test)

train_score_2 = r2_score(y_train_2, pred_train_2)
test_score_2 = r2_score(y_test_2, pred_test_2)


print("VIF data")
print("Train Score : %.4f" % train_score)
print("Test Score : %.4f" % test_score)


print("PCA data")
print("Train Score : %.4f" % train_score_2)
print("Test Score : %.4f" % test_score_2)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error
```

```python
param_grid = {
    "max_depth" : [8, 16, None],
    "n_estimators" : [100, 300, 500],
    "max_bin" : [20],
    "learning_rate" : [0.001, 0.0025, 0.003],
    "objective" : ["regression"],
    "boosting_type" : ["gbdt"],
    "metric" : ["mae"],
    "sub_feature" : [0.345],
    "bagging_fraction" : [0.7, 0.75, 0.85],
    "bagging_freq" : [40],
    "num_leaves" : [256, 512],
    "min_data" : [500],
    "verbose" : [-1], # 필수
    "min_hessian" : [0.05],
    "feature_fraction_seed" : [2],
    "bagging_seed" : [3]
}

gcv = GridSearchCV(estimator=model, param_grid=param_grid, cv=5,
                  n_jobs=-1, verbose=1)

gcv_2 = GridSearchCV(estimator=model_2, param_grid=param_grid, cv=5,
                  n_jobs=-1, verbose=1)

gcv.fit(X_train, y_train)
gcv_2.fit(X_train_2, y_train_2)
print("Best Estimator : ", gcv.best_estimator_)
print("Best Estimator : ", gcv_2.best_estimator_)
```

```python
print("Prediction with Best Estimator")
gcv_pred_train = gcv.predict(X_train)
gcv_pred_test = gcv.predict(X_test)

gcv_pred_train_2 = gcv_2.predict(X_train_2)
gcv_pred_test_2 = gcv_2.predict(X_test_2)

gcv_train_score = r2_score(y_train, gcv_pred_train)
gcv_test_score = r2_score(y_test, gcv_pred_test)

gcv_train_score_2 = r2_score(y_train_2, gcv_pred_train_2)
gcv_test_score_2 = r2_score(y_test_2, gcv_pred_test_2)


print("VIF data")
print("Train MAE Score : %.4f" % gcv_train_score)
print("Test MAE Score : %.4f" % gcv_test_score)

print("PCA data")
print("Train MAE Score : %.4f" % gcv_train_score_2)
print("Test MAE Score : %.4f" % gcv_test_score_2)
```

```python
print("Performance Gain") # 이걸로 성능 향상 확인.
print("VIF data")
print("in train : ", (train_score - gcv_train_score))
print("in test : ", (test_score - gcv_test_score))

print("PCA data")
print("in train : ", (train_score_2 - gcv_train_score_2))
print("in test : ", (test_score_2 - gcv_test_score_2))
```
